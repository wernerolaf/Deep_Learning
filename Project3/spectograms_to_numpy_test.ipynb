# test data set to numpy tensor

```python
import numpy as np
from PIL import Image
import os
import pandas as pd
import tensorflow as tf
```

```python
os.listdir("test/audio_spec/")
```

```python
X_data = []
files = os.listdir("test/audio_spec/")
i=0
for my_file in files:
    image = Image.open("test/audio_spec/"+my_file).convert('RGB')
    image = np.array(image)
    image=tf.image.rgb_to_grayscale(image).numpy()
    X_data.append(image)
    i+=1
    if i%1000==0:
        print(i)
```

```python
img_all=np.zeros([len(X_data),max([len(x) for x in X_data]),max([len(x[0]) for x in X_data]),1],dtype="uint8")
```

```python
for i in range(len(X_data)):
    img_all[i][0:len(X_data[i]),0:len(X_data[i][0])]=X_data[i]
```

```python
np.save("test_grey.npy",img_all)
```
