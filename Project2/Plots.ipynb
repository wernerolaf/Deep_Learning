```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df3l = pd.read_csv('logs3l.csv')
```

```python
loss_train = df3l['train_loss']
loss_val = df3l['valid_loss']
epochs = range(1,97)
plt.plot(epochs, loss_train, 'g', label='Training loss')
plt.plot(epochs, loss_val, 'b', label='validation loss')
plt.title('Training and Validation loss for 3-layer final ResNet')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
loss_train = df3l['train_acc']
loss_val = df3l['valid_acc']
epochs = range(1,97)
plt.plot(epochs, loss_train, 'g', label='Training accuracy')
plt.plot(epochs, loss_val, 'b', label='validation accuracy')
plt.title('Training and Validation accuracy for 3-layer final ResNet')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
