```python
import pandas as pd
import seaborn as sns
```

```python
history2=pd.read_csv("history_block1_width1_aug.csv")

history2["model"]="block1_augumentation"

history1=pd.read_csv("history_block1_width1_no_reg.csv")

history1["model"]="block1"

history3=pd.read_csv("history_block2_width1_aug.csv")

history3["model"]="block2_augumentation"

history4=pd.read_csv("history_block2_width1_aug_l2.csv")

history4["model"]="block2_augumentation_l2"

history5=pd.read_csv("history_block2_width1_reg_full.csv")

history5["model"]="block2_augumentation_l2_dropout"

history6=pd.read_csv("history_block3_width1_reg_full.csv")

history6["model"]="block3_augumentation_l2_dropout"

history7=pd.read_csv("history_block4_width1_reg_full.csv")

history7["model"]="block4_augumentation_l2_dropout"

history8=pd.read_csv("history_block5_width1_reg_full.csv")

history8["model"]="block5_augumentation_l2_dropout"
```

```python
history_all=pd.concat([history1,history2,history3,history4,history5,history6,history7,history8])
```

```python
history_all["Epoch"]=history_all["Unnamed: 0"]+1
```

```python
history_all["Loss_Diffrence"]=history_all.val_loss-history_all.loss
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
sns.lineplot(data=history_all,x="Epoch",y="Loss_Diffrence",hue="model").set(ylabel="Diffrence between Validation and Training Loss")
```

```python
sns.lineplot(data=history_all.loc[history_all.model!="block1"],x="Epoch",y="Loss_Diffrence",hue="model").set(ylabel="Diffrence between Validation and Training Loss")
```

```python
sns.lineplot(data=history_all,x="Epoch",y="val_acc",hue="model").set(ylabel="Validation Accuracy")
```

```python
mask=(history_all.model=="block2_augumentation_l2_dropout") | (history_all.model=="block2_augumentation_l2") | (history_all.model=="block2_augumentation")
```

```python
history_bl2=history_all.loc[mask]
```

```python
sns.lineplot(data=history_bl2,x="Epoch",y="val_acc",hue="model").set(ylabel="Validation Accuracy")
```

```python
sns.lineplot(data=history_bl2,x="Epoch",y="Loss_Diffrence",hue="model").set(ylabel="Diffrence between Validation and Training Loss")
```

```python
sns.lineplot(data=history_all.loc[(history_all.Epoch>20) & (history_all.model!="block1") & (history_all.model!="block1_augumentation")],x="Epoch",y="val_acc",hue="model").set(ylabel="Validation Accuracy")
```
